```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Nodes
nodes = {
    "Input": (0, 6),
    "1D Conv": (0, 5),
    "Max Pooling": (1, 5),
    "Dropout_1": (2, 5),
    "BiLSTM_1": (0, 4),
    "BiLSTM_2": (1, 4),
    "BiLSTM_3": (2, 4),
    "Dropout_2": (3, 4),
    "Dense_1": (0, 3),
    "Dense_2": (1, 3),
    "Dense_3": (2, 3),
    "Dropout_3": (3, 3),
    "Output": (1, 2)
}

# Edges
edges = [
    ("Input", "1D Conv"),
    ("1D Conv", "Max Pooling"),
    ("Max Pooling", "Dropout_1"),
    ("Dropout_1", "BiLSTM_1"),
    ("BiLSTM_1", "BiLSTM_2"),
    ("BiLSTM_2", "BiLSTM_3"),
    ("BiLSTM_3", "Dropout_2"),
    ("Dropout_2", "Dense_1"),
    ("Dense_1", "Dense_2"),
    ("Dense_2", "Dense_3"),
    ("Dense_3", "Dropout_3"),
    ("Dropout_3", "Output")
]

# Add nodes and edges to graph
for node, pos in nodes.items():
    G.add_node(node, pos=pos)

for edge in edges:
    G.add_edge(*edge)

# Get positions
pos = {node: (x, y) for node, (x, y) in nodes.items()}

# Define node colors
node_colors = {
    "1D Conv": "orange",
    "Max Pooling": "khaki",
    "Dropout_1": "lightcoral",
    "BiLSTM_1": "lightblue",
    "BiLSTM_2": "lightblue",
    "BiLSTM_3": "lightblue",
    "Dropout_2": "lightcoral",
    "Dense_1": "lightgreen",
    "Dense_2": "lightgreen",
    "Dense_3": "lightgreen",
    "Dropout_3": "lightcoral",
    "Input": "white",
    "Output": "white"
}

# Draw graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color=[node_colors.get(node, "white") for node in G.nodes],
        font_size=10, edge_color="black", font_weight="bold", linewidths=1.5)

# Show plot
plt.title("Neural Network Architecture with Dropout")
plt.show()
```
